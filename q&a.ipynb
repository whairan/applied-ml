```python
def find_cat_correlation(data, cat_cols, target_col):
    """
    Finds the categorical column most highly correlated (using Cramér's V) with the target column.

    Args:
        data: Pandas DataFrame.
        cat_cols: List of categorical columns.
        target_col: Name of the target column (must be categorical).

    Returns:
        Name of the most highly correlated categorical column.
    """
    from scipy.stats import chi2_contingency
    import numpy as np

    correlations = {}

    for col in cat_cols:
        if col != target_col:
            # Create a contingency table
            contingency_table = pd.crosstab(data[col], data[target_col])

            # Perform chi-squared test
            chi2, _, _, _ = chi2_contingency(contingency_table)

            # Calculate Cramér's V
            n = contingency_table.sum().sum()  # Total number of observations
            phi2 = chi2 / n  # Calculate phi-squared
            min_dim = min(contingency_table.shape) - 1  # Adjust for degrees of freedom
            cramers_v = np.sqrt(phi2 / min_dim)  # Cramér's V formula

            # Store the correlation
            correlations[col] = cramers_v

    # Find the column with the highest correlation
    most_correlated_column = max(correlations, key=correlations.get)
    return most_correlated_column

# Find the most correlated column
most_correlated_cat_col = find_cat_correlation(
    titanic_data, ['sex', 'embarked', 'class', 'who'], 'pclass'
)

print("Most highly correlated categorical column with pclass:", most_correlated_cat_col)
```
